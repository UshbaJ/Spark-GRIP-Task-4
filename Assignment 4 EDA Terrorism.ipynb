### Heat map of Covariance among the set of variables

```python
#Performing Covariance
# Select only the numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix
covariance_matrix = numeric_df.cov()

covariance_matrix
```

### Number of Terrorist  activities by region

```python
plt.subplots(figsize = (15,16))
sns.countplot(x = 'region_txtt', data = df, palette = 'inferno', order = df['region_txtt'].value_counts().index)
plt.xticks(rotation = 90)
plt.xlabel('Regions',fontsize = 14)
plt.title('Terrorist activities by region',fontsize = 22)
plt.show()
```

###  Number of Terrorist Acitvities by each Year

```python
plt.subplots(figsize = (16,8))
sns.countplot(x = 'uyear',data = df, palette = 'RdYlGn_r', edgecolor = sns.color_palette('dark',8))
plt.xticks(rotation = 90)
plt.xlabel('Year of attack',fontsize = 14)
plt.title('Number of Terrorist Activities by Each Year',fontsize = 22)
plt.show()
```

### Frequency of Attacks

```python
plt.subplots(figsize = (18,8))
sns.countplot(x = df['uyear'], hue = 'success', data = df)
plt.xticks(rotation = 90)
plt.xlabel('Years',fontsize = 14)
plt.title('Frequency of Attacks',fontsize = 22)
plt.show()
```

### Attack Overview

```python
count_terror = df['country_txtt'].value_counts()[:15].to_frame()
count_terror.columns = ['Attack']
count_kill = df.groupby('country_txtt')['nkilledter'].sum().to_frame()
count_terror.merge(count_kill, left_index = True, right_index = True, how = 'left').plot.bar(width = 0.9)
plt.xlabel('Countries', fontsize = 14)
plt.title('Attack Stats', fontsize = 22)
fig = plt.gcf()
fig.set_size_inches(16,5)
plt.show()
```
